Ejercicio 3

Usando "information_schema", extraer los nombres de las columnas de la tabla "customers" de la BBDD "northwind".

```python
import mysql.connector

#aqui creo el conector 
cnx = mysql.connector.connect(user="root", 
                              password="AlumnaAdalab", 
                              host="localhost", 
                              database="northwind")

#aqui creo el cursor 
mycursor = cnx.cursor()
query_customers = "SELECT column_name FROM information_schema.columns WHERE table_schema = 'northwind' AND table_name = 'customers'"

#aqui ejecuto la query que he pedido 
mycursor.execute(query_customers)

nombre_columnas = mycursor.fetchall()

for columna in nombre_columnas:
    print(columna[0])



mycursor.close()
cnx.close()
```

```python
df = pd.read_sql_query(nombre_columnas, cnx)
```
